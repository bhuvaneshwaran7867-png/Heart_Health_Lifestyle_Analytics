```python
# Create BMI Category
care["BMI_Category"] = pd.cut(
    care["BMI"],
    bins=[0, 18.5, 24.9, 29.9, float("inf")],
    labels=["Underweight", "Healthy Weight", "Overweight", "Obese"],
    include_lowest=True
)
```

```python
import numpy as np

care["BMI_Category"] = np.select(
    [
        care["BMI"] < 18.5,
        (care["BMI"] >= 18.5) & (care["BMI"] < 25),
        (care["BMI"] >= 25) & (care["BMI"] < 30),
        care["BMI"] >= 30
    ],
    [
        "Underweight",
        "Healthy Weight",
        "Overweight",
        "Obese"
    ],
    default="Unknown"
)

print(care["BMI_Category"].value_counts())
```

# Plot 1: Box Plot (Clinical Metric vs. Discrete Target)

```python
df_numeric_eval = care.copy()
df_numeric_eval['Heart_Attack_Flag'] = df_numeric_eval['Heart_Attack'].map({'Yes': 1, 'No': 0})

core_metrics = [
    'Age', 'BMI', 'Blood_Pressure_Systolic', 'Cholesterol', 'Blood_Sugar',
    'Exercise_Hours_Per_Week', 'Daily_Steps', 'Sleep_Hours', 'Stress_Level', 
    'Resting_Heart_Rate', 'Heart_Attack_Flag'
]
```

```python
# Create two plots side by side
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Box Plot: Blood Pressure vs Heart Attack
sns.boxplot(
    data=care,
    x="Heart_Attack",
    y="Blood_Pressure_Systolic",
    ax=axes[0]
)

axes[0].set_title("Blood Pressure vs Heart Attack")
axes[0].set_xlabel("Heart Attack")
axes[0].set_ylabel("Systolic Blood Pressure")

# Scatter Plot: Exercise Hours vs Daily Steps
sns.scatterplot(
    data=care,
    x="Exercise_Hours_Per_Week",
    y="Daily_Steps",
    hue="Heart_Attack",
    ax=axes[1]
)

axes[1].set_title("Exercise Hours vs Daily Steps")
axes[1].set_xlabel("Exercise Hours Per Week")
axes[1].set_ylabel("Daily Steps")
axes[1].legend(title="Heart Attack")

plt.tight_layout()
plt.show()
```

```python
# Plot 3: Heatmap (System-Wide Correlations)
fig, axes = plt.subplots(1, 1, figsize=(18, 6))

corr_matrix = df_numeric_eval[core_metrics].corr()

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    ax=axes
)

axes.set_title("System Correlation Matrix")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python

# Load the dataset
care = pd.read_excel(
    r"C:\Users\user\Pictures\Desktop\Heart_Health_Lifestyle_Analytics\Dataset\Heart_Health_Lifestyle_Analytics_2000_Records.xlsx", 
    sheet_name="Heart_Health_Data"
)

# Clean the data
care = care.drop_duplicates(subset="Patient_ID")
care["Smoking"] = care["Smoking"].replace("Y", "Yes")
care["City"] = care["City"].str.strip().str.title()

# Fill missing values
care["BMI"] = care["BMI"].fillna(care["BMI"].median())
care["Sleep_Hours"] = care["Sleep_Hours"].fillna(care["Sleep_Hours"].median())
care["Hospital_Visits_Last_Year"] = care["Hospital_Visits_Last_Year"].fillna(
    care["Hospital_Visits_Last_Year"].median()
)

# Remove invalid ages
care = care[care["Age"] < 100]

# Create a numeric column for correlation
care["Heart_Attack_Flag"] = care["Heart_Attack"].map({"Yes": 1, "No": 0})

# Columns used for the correlation heatmap
columns = [
    "Age",
    "BMI",
    "Blood_Pressure_Systolic",
    "Cholesterol",
    "Blood_Sugar",
    "Exercise_Hours_Per_Week",
    "Daily_Steps",
    "Sleep_Hours",
    "Stress_Level",
    "Resting_Heart_Rate",
    "Heart_Attack_Flag"
]

# Create the dashboard
fig = plt.figure(figsize=(20, 10))
grid = plt.GridSpec(2, 3)

# Age Distribution
ax1 = fig.add_subplot(grid[0, 0])

sns.histplot(data=care, x="Age", bins=15, kde=True, ax=ax1)

ax1.set_title("Age Distribution")
ax1.set_xlabel("Age")
ax1.set_ylabel("Count")

# Blood Pressure Distribution
ax2 = fig.add_subplot(grid[0, 1])

sns.histplot(
    data=care,
    x="Blood_Pressure_Systolic",
    kde=True,
    ax=ax2
)

ax2.set_title("Blood Pressure Distribution")
ax2.set_xlabel("Systolic Blood Pressure")
ax2.set_ylabel("Count")

# Correlation Heatmap
ax3 = fig.add_subplot(grid[0, 2])

corr = care[columns].corr()

sns.heatmap(corr, cmap="coolwarm", ax=ax3)

ax3.set_title("Correlation Matrix")

# Exercise vs Daily Steps
ax4 = fig.add_subplot(grid[1, :2])

sns.scatterplot(
    data=care,
    x="Exercise_Hours_Per_Week",
    y="Daily_Steps",
    hue="Heart_Attack",
    ax=ax4
)

ax4.set_title("Exercise Hours vs Daily Steps")
ax4.set_xlabel("Exercise Hours Per Week")
ax4.set_ylabel("Daily Steps")

# Family History vs Smoking
ax5 = fig.add_subplot(grid[1, 2])

sns.countplot(
    data=care,
    x="Family_History",
    hue="Smoking",
    ax=ax5
)

ax5.set_title("Family History and Smoking")
ax5.set_xlabel("Family History")
ax5.set_ylabel("Count")

# Dashboard title
plt.suptitle("Heart Health Dashboard", fontsize=16)

plt.tight_layout()
plt.show()
```
